# load example data

```python
import numpy as np
from monty.serialization import loadfn

compositions, temperature, adsorptions = loadfn('plot_data.json')

X = np.stack((compositions, temperature), axis=1)
y = np.array(adsorptions)[:, np.newaxis]
```

# Initialize phase diagram plotter

```python
from gbdnn.plotter import Plotter

plotter = Plotter()
plotter.smooth(X, y)
```

# Show projected interpolated complexion diagram

```python
plt_flat = plotter.get_plot(formatter='flat', vmin=0, vmax=25)
plt_flat.show()
```

# Show 3-d interpolated phase diagram

```python
plt_surf = plotter.get_plot(formatter='surface')
plt_surf.show()
```
